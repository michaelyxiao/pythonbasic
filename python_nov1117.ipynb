# Loading Data with Pandas

```python
# This is a comment. You can add a comment by adding a <#> in front of a line.
# read_csv is a pandas function that takes a comma-separated file and converts it into a pandas dataframe
# in memory (i.e. loaded into your computer's RAM).
# A pandas dataframe is an extremely useful data structure that has a large number of built-in
# data manipulation and data analysis functions.

df = pd.read_csv('C:/Users/Michael/Desktop/multipleChoiceResponses.csv', encoding="ISO-8859-1", low_memory=False,
                thousands = ',')
schema = pd.read_csv('C:/Users/Michael/Desktop/schema.csv')
```

```python
# One of the first things we want to check is the number of nulls in a dataset.
# The isnull() function of a dataframe returns a True/False value indicating whether
# a value is null or not.

df.isnull()
```

```python
# A dataframe is useful, but we really want to know how many nulls there are
# for each column. This will help us decide what columns to drop.

df.isnull().sum()
```

```python
# The returned list above is cutoff -- so we can force pandas to display all rows.

pd.set_option('display.max_rows', None)
df.isnull().sum()
```

```python
# Lists in Python are captured in brackets: [1, 2, 3]. To create a new dataframe
# with a subset of columns of another dataframe, we can specify it in the manner
# below.

df_small = df[['Age', 'Country', 'EmploymentStatus']]
df_small.head()
```

```python
# Since there are a lot of questions that went unanswered with a high volume
# of nulls, we're going to want to drop many of them. I arbitrarily selected
# a value of 25% answered as the cut-off.
#
# The dropna function drops null columns from a dataframe. axis = 1 specifies
# column (axis = 0 specifies index, not row), and the len function returns
# the length of any datatype in Python (in this case it returns the number of
# rows in the dataframe df).

df_new = df.dropna(axis = 1, thresh = len(df) * .25)
df_new.head()
```

```python
# The value_counts function also has a nlargest function that returns
# the n largest values.

df_new['Country'].value_counts().nlargest(10)
```

```python
# For a bar chart, you need to specify how many unique x values there are.
# The numpy arange function returns a list from 0 to 9 (if 10 is the parameter)
x = np.arange(10)

# clf clears the matplotlib figure -- otherwise you'll end up stacking plots
# one on top of the other.
plt.clf()
# Sets the figure size -- note the numbers are in inches.
plt.figure(figsize=(30,15))
# Creates a bar chart with x values, and the heights are equal to the number
# of respondents form each of the top 10 countries.
plt.bar(x, df_new['Country'].value_counts().nlargest(10).values)
# The index function returns the actual names associated with the top 10 
# countries.
plt.xticks(x, df_new['Country'].value_counts().nlargest(10).index, rotation = 45)

plt.title("Survey Respondents by Country", fontsize = 30)
plt.xlabel("Country", fontsize = 24)
plt.ylabel("Number of Respondents", fontsize = 24)

plt.grid(False)

# This sets the x/y label font sizes to be larger. Note that this is global
# and will apply to all future plots.
matplotlib.rc('xtick', labelsize=24) 
matplotlib.rc('ytick', labelsize=24) 

plt.show()
```

```python
# How about a horizontal bar chart? Just flip everything around and use barh
# instead of bar

y = np.arange(10)

plt.clf()
plt.figure(figsize=(30,15))
plt.barh(y, df_new['Country'].value_counts().nlargest(10).values)
plt.yticks(y, df_new['Country'].value_counts().nlargest(10).index, rotation = 45)

plt.title("Survey Respondents by Country", fontsize = 30)
plt.ylabel("Country", fontsize = 24)
plt.xlabel("Number of Respondents", fontsize = 24)

plt.grid(False)

plt.show()
```

```python
# We only care about the US...
# Also people who actually code for their job.

df_final = df_new[(df_new.Country == 'United States') &
                 (df_new.Age > 21) &
                 (df_new.Age <= 65) &
                 (df_new.CodeWriter == 'Yes')]
len(df_final)
```

```python
# Have to clean up the compensation column to remove commas.
#
# Note the warning is due to chained assignment, which can create confusion around what value
# actually gets assigned when there are multiple steps, but we can ignore for now.

df_final['CompensationAmount'] = df_final['CompensationAmount'].astype(str).str.replace(',','')
df_final = df_final[(df_final['CompensationAmount'] != 'nan')]
df_final['CompensationAmount'] = df_final['CompensationAmount'].astype(float)
df_final = df_final[(df_final['CompensationAmount'] > 10000) &
                                        (df_final['CompensationAmount'] < 1000000)]
```

```python
df_final['CompensationAmount']
```

```python
import matplotlib.patches as mpatches

# Let's make a histogram of compensation

plt.clf()
plt.figure(figsize=(30,15))
plt.hist(df_final['CompensationAmount'], rwidth = .9, bins = 40)

# We can draw lines directly on the plot -- here we draw the median in red and
# mean in green
plt.axvline(df_final['CompensationAmount'].median(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(df_final['CompensationAmount'].mean(), color='g', linestyle='dashed', linewidth=2)

plt.title("Compensation (Salary + Bonus)", fontsize = 30)
plt.ylabel("Respondents", fontsize = 24)
plt.xlabel("Compensation", fontsize = 24)
plt.xticks([50000, 100000, 150000, 200000, 250000, 300000, 350000, 400000, 450000, 500000, 550000])

# Add a legend
plt.rcParams["legend.fontsize"] = 30
red_patch = mpatches.Patch(color='red', label='Median')
green_patch = mpatches.Patch(color='green', label='Average')
plt.legend(handles=[red_patch, green_patch])

plt.show()
```

```python
# Let's do a line chart

plt.clf()
plt.figure(figsize=(30,15))
plt.plot(df_final['CompensationAmount'].groupby(df_final['Age']).mean().index, 
         df_final['CompensationAmount'].groupby(df_final['Age']).mean(),
        linewidth = 10)
plt.yticks([50000, 75000, 100000, 125000, 150000, 175000, 200000])

plt.title("Average Compensation by Age", fontsize = 30)
plt.ylabel("Salary", fontsize = 24)
plt.xlabel("Age", fontsize = 24)

plt.grid(False)

plt.show()
```

```python
# Seaborn (sns) is built on top of matplotlib and providers
# more colorful visualizations
#
# Violin chart

plt.clf()
plt.figure(figsize=(30,15))
sns.violinplot(y='CompensationAmount', x='FormalEducation', data=df_final, bw = .5, legend = True)
plt.title("Salary distribution for US data scientists", fontsize=24)
plt.ylabel("Annual Salary", fontsize=24)
plt.xlabel("Education Level", fontsize=24)
plt.show()
```

```python
# Colorful bar chart
plt.clf()
plt.figure(figsize=(30,15))
sns.countplot(y=df_final['MajorSelect'])
plt.title("College Majors", fontsize=30)
plt.xlabel("Number of respondents", fontsize=24)
plt.ylabel("Major", fontsize=24)
plt.show()
```
